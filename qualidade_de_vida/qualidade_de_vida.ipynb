```python
# Verificando valores ausentes
print("\nValores ausentes por coluna:")
df.isnull().sum()
```

```python
# Limpando colunas com inconsistências e convertendo colunas numéricas armazenadas como strings
cols_to_clean = ['Property Price to Income Value', 'Quality of Life Value']
for col in cols_to_clean:
    df[col] = df[col].replace({":": "", "'": ""}, regex=True).replace("None", np.nan).astype(float)
```

```python
# Exibindo dados limpos
print("\nDados após limpeza:")
df.head()
```

### Plotagem

```python
# Visualizações iniciais
plt.figure(figsize = (10, 6))
sns.histplot(df['Quality of Life Value'], kde = True, bins = 20, color = 'blue')
plt.title('Distribuição da Qualidade de Vida')
plt.xlabel('Quality of Life Value')
plt.ylabel('Frequência')
plt.show()
```

```python
# Converte a coluna 'Property Price to Income Value' para valores numéricos
df['Property Price to Income Value'] = pd.to_numeric(df['Property Price to Income Value'], errors = 'coerce')

# Cálculo da mediana para preenchimento de valores ausentes
median_val = df['Property Price to Income Value'].median()

# Preenchimento dos valores ausentes com a mediana
df['Property Price to Income Value'].fillna(median_val, inplace = True)
```

```python
num_col = [col for col in df.columns if df[col].dtype != 'object']
```

```python
clean_df = df.dropna(subset = num_col, how = 'all')
```

### Inferências

```python
# Relação entre "Safety Value" e "Quality of Life Value"
plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'Safety Value', y = 'Quality of Life Value', data = df, hue = 'Safety Category', palette = 'viridis')
plt.title('Relação entre Segurança e Qualidade de Vida')
plt.xlabel('Safety Value')
plt.ylabel('Quality of Life Value')
plt.show()
```

### Analisando a influência do custo de vida na qualidade de vida

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x = 'Cost of Living Category', y = 'Quality of Life Value', data = df, palette = 'coolwarm')
plt.title('Custo de Vida vs Qualidade de Vida')
plt.xlabel('Categoria de Custo de Vida')
plt.ylabel('Qualidade de Vida')
plt.show()
```

### Análise adicional: Distribuição por continente

```python
if 'Continent' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x = 'Continent', y = 'Quality of Life Value', data = df, palette = 'Set2')
    plt.title('Distribuição da Qualidade de Vida por Continente')
    plt.xlabel('Continente')
    plt.ylabel('Qualidade de Vida')
    plt.xticks(rotation = 45)
    plt.show()
```

### Análise adicional: Correlações

```python
plt.figure(figsize = (12, 8))
corr_matrix = df.select_dtypes(include = np.number).corr()
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt = '.2f', linewidths = 0.5)
plt.title('Mapa de Correlação entre as Variáveis')
plt.show()
```

### Análise adicional: Impacto do transporte público

```python
if 'Traffic Commute Time Value' in df.columns:
    plt.figure(figsize = (10, 6))
    sns.scatterplot(x = 'Traffic Commute Time Value', y = 'Quality of Life Value', data = df, color='orange')
    plt.title('Impacto do Tempo de Transporte Público na Qualidade de Vida')
    plt.xlabel('Traffic Commute Time Value')
    plt.ylabel('Quality of Life Value')
    plt.show()
```

### Analisando os países com melhor qualidade de vida

```python
top_countries = df[['country', 'Quality of Life Value']].sort_values(by = 'Quality of Life Value', ascending = False).head(10)
print("\nTop 10 países com maior qualidade de vida:")
top_countries
```

### Comparação Brasil e Alemanha

```python
brazil_germany = df[df['country'].isin(['Brazil', 'Germany'])]
```

```python
# Visualização
metrics = ['Quality of Life Value', 'Safety Value', 'Cost of Living Value', 'Traffic Commute Time Value']
```

```python
for metric in metrics:
    plt.figure(figsize = (8, 5))
    sns.barplot(x = 'country', y = metric, data = brazil_germany, palette = 'Set3')
    plt.title(f'Comparação de {metric} entre Brasil e Alemanha')
    plt.xlabel('País')
    plt.ylabel(metric)
    plt.show()
```
